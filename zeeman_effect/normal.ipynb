```python
import numpy as np
import scipy
from matplotlib import pyplot as plt
import matplotlib as mpl
import csv
```

```python
pxc = []
ic = []
with open('calib.dat') as data:
    csv_reader = csv.reader(data, delimiter=' ')
    
    for line in csv_reader:
        pxc.append(int(line[0]))
        ic.append(int(line[1]))
        
```

```python
pxc = np.array(pxc)
ic = np.array(ic)
```

```python
plt.plot(pxc,ic)
plt.xlabel('Pixel')
plt.ylabel('Relative intensity')

plt.show()
```
