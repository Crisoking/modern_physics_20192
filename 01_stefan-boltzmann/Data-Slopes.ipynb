```python
def sb_law (T, A, B, a=4, b=1):
    return (A*(T**a-(294.6)**a)+B*(T-294.6)**b)
```

```python
popt, pcov = curve_fit(sb_law, temp, power, method='lm')
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,1,1)
plt.plot(temp,power,'k')
plt.plot(x,sb_law(x,popt[0],popt[1]), 'r:')
plt.xlabel('Temperature (K)')
plt.ylabel('Power (mW)')
plt.subplot(2,1,2)
plt.scatter(temp,sb_law(temp,popt[0],popt[1])-v*i)
plt.xlabel('Temperature (K)')
plt.ylabel('Power residual (mW)')

plt.tight_layout()
plt.show()
```

## Error propagation

```python
unc_pow = np.array([0.02141, 0.1837, 0.2241, 0.6330000000000001, 1.2109999999999999, 1.493, 2.4050000000000002, 10.22,
           12.600000000000001, 14.86, 19.38, 24.400000000000002, 28.610000000000003])
```

The formula for the uncertainty of the slope of a data set fitted to $y=A+Bx$ is given by: $$\sigma_B=\sigma_y\sqrt{\frac{N}{\Delta}} $$
where
$$\sigma_y=\sqrt{\frac{1}{N-2}\sum_{i=1}^N(y_i-A-Bx_i)^2} $$
$$\Delta = N\sum x^2-\left(\sum x^2\right)$$
and $N$ is the size of the data set.


```python
A=[np.polyfit(I_1,V_1,1)[1],np.polyfit(I_2,V_2,1)[1],np.polyfit(I_3,V_3,1)[1],np.polyfit(I_4,V_4,1)[1],
  np.polyfit(I_5,V_5,1)[1],np.polyfit(I_6,V_6,1)[1],np.polyfit(I_7,V_7,1)[1],np.polyfit(I_8,V_8,1)[1],
  np.polyfit(I_9,V_9,1)[1],np.polyfit(I_10,V_10,1)[1],np.polyfit(I_11,V_11,1)[1],np.polyfit(I_12,V_12,1)[1],
  np.polyfit(I_13,V_13,1)[1]]
N=[V_1.size,V_2.size,V_3.size,V_4.size,V_5.size,V_6.size,V_7.size,V_8.size,V_9.size,V_10.size,V_11.size,V_12.size,
   V_13.size]
```

```python
sum_1=0
for n in range(V_1.size):
    sum_1+=(V_1[n]-A[0]-R[0]*I_1[n])**2
sum_2=0
for n in range(V_2.size):
    sum_2+=(V_2[n]-A[1]-R[1]*I_2[n])**2
sum_3=0
for n in range(V_3.size):
    sum_3+=(V_3[n]-A[2]-R[2]*I_3[n])**2
sum_4=0
for n in range(V_4.size):
    sum_4+=(V_4[n]-A[3]-R[3]*I_4[n])**2
sum_5=0
for n in range(V_5.size):
    sum_5+=(V_5[n]-A[4]-R[4]*I_5[n])**2
sum_6=0
for n in range(V_6.size):
    sum_6+=(V_6[n]-A[5]-R[5]*I_6[n])**2
sum_7=0
for n in range(V_7.size):
    sum_7+=(V_7[n]-A[6]-R[6]*I_7[n])**2
sum_8=0
for n in range(V_8.size):
    sum_8+=(V_8[n]-A[7]-R[7]*I_8[n])**2
sum_9=0
for n in range(V_9.size):
    sum_9+=(V_9[n]-A[8]-R[8]*I_9[n])**2
sum_10=0
for n in range(V_10.size):
    sum_10+=(V_10[n]-A[9]-R[9]*I_10[n])**2
sum_11=0
for n in range(V_11.size):
    sum_11+=(V_11[n]-A[10]-R[10]*I_11[n])**2
sum_12=0
for n in range(V_12.size):
    sum_12+=(V_12[n]-A[11]-R[11]*I_12[n])**2
sum_13=0
for n in range(V_13.size):
    sum_13+=(V_13[n]-A[12]-R[12]*I_13[n])**2
```

```python
Delta_1=I_1.size*(sum(I_1**2))-(sum(I_1))**2
Delta_2=I_2.size*(sum(I_2**2))-(sum(I_2))**2
Delta_3=I_3.size*(sum(I_3**2))-(sum(I_3))**2
Delta_4=I_4.size*(sum(I_4**2))-(sum(I_4))**2
Delta_5=I_5.size*(sum(I_5**2))-(sum(I_5))**2
Delta_6=I_6.size*(sum(I_6**2))-(sum(I_6))**2
Delta_7=I_7.size*(sum(I_7**2))-(sum(I_7))**2
Delta_8=I_8.size*(sum(I_8**2))-(sum(I_8))**2
Delta_9=I_9.size*(sum(I_9**2))-(sum(I_9))**2
Delta_10=I_10.size*(sum(I_10**2))-(sum(I_10))**2
Delta_11=I_11.size*(sum(I_11**2))-(sum(I_11))**2
Delta_12=I_12.size*(sum(I_12**2))-(sum(I_12))**2
Delta_13=I_13.size*(sum(I_13**2))-(sum(I_13))**2
D=np.array([Delta_1,Delta_2,Delta_3,Delta_4,Delta_5,Delta_6,Delta_7,Delta_8,Delta_9,Delta_10,Delta_11,Delta_12,Delta_13])
```

```python
sig_1=np.sqrt(1/(V_1.size-2)*sum_1)
sig_2=np.sqrt(1/(V_2.size-2)*sum_2)
sig_3=np.sqrt(1/(V_3.size-2)*sum_3)
sig_4=np.sqrt(1/(V_4.size-2)*sum_4)
sig_5=np.sqrt(1/(V_5.size-2)*sum_5)
sig_6=np.sqrt(1/(V_6.size-2)*sum_6)
sig_7=np.sqrt(1/(V_7.size-2)*sum_7)
sig_8=np.sqrt(1/(V_8.size-2)*sum_8)
sig_9=np.sqrt(1/(V_9.size-2)*sum_9)
sig_10=np.sqrt(1/(V_10.size-2)*sum_10)
sig_11=np.sqrt(1/(V_11.size-2)*sum_11)
sig_12=np.sqrt(1/(V_12.size-2)*sum_12)
sig_13=np.sqrt(1/(V_13.size-2)*sum_13)
sigma = np.array([sig_1,sig_2,sig_3,sig_4,sig_5,sig_6,sig_7,sig_8,sig_9,sig_10,sig_11,sig_12,sig_13])
```

```python
unc_r = []
for n in range(13):
    unc_r.append(sigma[n]*np.sqrt(N[n]/D[n]))
```

```python
unc_rest = ((np.true_divide(unc_r,R))+(0.001/R_0))*rest
```
