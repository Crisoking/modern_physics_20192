```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import scipy
# from scipy import linalg as lin
# from scipy import stats
import scipy.interpolate as intrp
from matplotlib import rc
# import sys
# import re
# import pandas as pd
temp=np.linspace(300,3600,num = 34)
rest=np.array([5.65,8.06,10.56,13.23,16.09,19,21.94,24.93,27.94,30.98,34.08,37.19,40.36,43.55,46.78,50.05,
              53.35,56.67,60.06,63.48,66.91,70.39,73.91,77.49,81.04,84.7,88.33,92.04,95.76,99.54,103.3,
              107.2,111.1,115.0])
```

```python
z= np.polyfit(rest,temp,7)
f=np.poly1d(z)
z= np.polyfit(temp,rest,7)
finv=np.poly1d(z)
temp1=np.linspace(250,3800,num = 60)
tempt=np.array([f(x) for x in rest])
rest1=np.array([finv(x) for x in temp])
```

```python
plt.subplot(1,2,1)
plt.plot(rest,temp,'g^')
plt.plot(rest,tempt,'r')
plt.ylabel('Temperatura (K)')
plt.xlabel(r'Resistividad ($\times 10^{-8} \Omega\cdot$m)')

plt.subplot(1,2,2)
plt.plot(rest,tempt-temp,'o')
plt.xlabel('Resistividad')
plt.ylabel('Residual de Temperatura (K)')
plt.tight_layout

plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(temp,rest, 'D')
plt.plot(temp,rest1, 'r')

plt.subplot(1,2,2)
plt.plot(temp,rest-rest1, 'o')
plt.show()
```
