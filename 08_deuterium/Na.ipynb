```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import scipy
import csv
from scipy.optimize import curve_fit
```

```python
px = []
i = []

with open('na.dat') as data:
    csv_reader = csv.reader(data, delimiter=' ')

    for line in csv_reader:
        px.append(float(line[0]))
        i.append(float(line[1]))

px = np.array(px)
i = np.array(i)
```

```python
plt.plot(px,i)
plt.xlabel('Pixels')
plt.ylabel('Relative intensity')
plt.show()
```

```python
s = 592.135-588.995
```

```python
np.where(i==max(i))
```

```python
np.where(i==max(i[150:200]))
```

```python
sc = s/(199-18)
```

```python
b = 588.995-18*sc
```

```python
wl = px*sc+b
```

```python
plt.figure(figsize=(3.404,3.404/1.61))
plt.plot(wl,i)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Relative intensity')
plt.tight_layout()
plt.savefig('na.pdf')
```
